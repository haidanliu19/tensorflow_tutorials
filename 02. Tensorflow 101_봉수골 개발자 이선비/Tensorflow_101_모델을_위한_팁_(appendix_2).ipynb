```python
import tensorflow as tf
import pandas as pd
```

```python
filePath = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv'
boston = pd.read_csv(filePath)
boston.head()
```

```python
X = boston[['crim', 'zn', 'indus', 'chas', 'nox', 
            'rm', 'age', 'dis', 'rad', 'tax',
            'ptratio', 'b', 'lstat']]
y = boston[['medv']]
print(X.shape, y.shape)
```

```python
X_input = tf.keras.layers.Input(shape=[13])
H = tf.keras.layers.Dense(8, activation='swish')(X_input)
H = tf.keras.layers.Dense(8, activation='swish')(H)
H = tf.keras.layers.Dense(8, activation='swish')(H)
Y_output = tf.keras.layers.Dense(1)(H)
model = tf.keras.models.Model(X_input, Y_output)
model.compile(loss='mse')
```

```python
X_input = tf.keras.layers.Input(shape=[13])

H = tf.keras.layers.Dense(8)(X_input)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dense(8)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dense(8)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

Y_output = tf.keras.layers.Dense(1)(H)
model = tf.keras.models.Model(X_input, Y_output)
model.compile(loss='mse')
```

```python
model.fit(X, y, epochs=1000)
```
