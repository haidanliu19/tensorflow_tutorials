```python
import pandas as pd
import tensorflow as tf
```

# 1. 데이터 준비

```python
lemonade = pd.read_csv('lemonade.csv')
lemonade
```

```python
boston = pd.read_csv('boston.csv')
boston
```

```python
iris = pd.read_csv('iris.csv')
iris
```

```python
print(lemonade.shape)
print(boston.shape)
print(iris.shape)
```

```python
# 컬럼이름 출력
print(lemonade.columns)
print(boston.columns)
print(iris.columns)
```

```python
iris_encoding = pd.get_dummies(iris)
iris_encoding.head()
```

```python
X = iris_encoding[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
y = iris_encoding[['품종_setosa', '품종_versicolor', '품종_virginica']]
print(X.shape, y.shape)
```

```python
iris_encoding.head()
```

# 모델의 구조

```python
X_input = tf.keras.layers.Input(shape=[4])
H = tf.keras.layers.Dense(8, activation="swish")(X_input)
H = tf.keras.layers.Dense(8, activation="swish")(H)
H = tf.keras.layers.Dense(8, activation="swish")(H)
Y_output = tf.keras.layers.Dense(3, activation='softmax')(H)
model = tf.keras.models.Model(X_input, Y_output)
model.compile(loss='categorical_crossentropy',
              metrics='accuracy')
```

```python
model.summary()
```

# 모델의 학습

```python
#model.fit(독립, 종속, epochs=1000, verbose=0)
model.fit(X, y, epochs=100)
```

# 모델의 이용

```python
# 4. 모델을 이용합니다
print(model.predict(X[0:5]))
# 종속변수 확인
print(y[0:5])
```

```python
# 모델의 수식 확인
print(model.get_weights())
```
